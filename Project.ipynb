Removing noise

```python
#Remove hashtags
df['clean_tweet'] = df['tweet'].apply(nfx.remove_hashtags)
```

```python
df[['tweet', 'clean_tweet']]
```

```python
#Remove handles
df['clean_tweet'] = df['clean_tweet'].apply(nfx.remove_userhandles)
```

```python
df[['tweet', 'clean_tweet']]
```

```python
#Remove multiple spaces
df['clean_tweet'] = df['clean_tweet'].apply(nfx.remove_multiple_spaces)
```

```python
df[['tweet', 'clean_tweet']]
```

```python
#Remove urls
df['clean_tweet'] = df['clean_tweet'].apply(nfx.remove_urls)
```

```python
#Remove punctuations
df['clean_tweet'] = df['clean_tweet'].apply(nfx.remove_puncts)
```

```python
df[['tweet', 'clean_tweet']]
```

```python
#removing words with less than 3 characters
df['clean_tweet'] = df['clean_tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3 ]))
df['clean_tweet'].head()
```

```python
#Removing stopwords
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
df['clean_tweet'] = df['clean_tweet'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df['clean_tweet']
```

```python
tokenized_tweet = df['clean_tweet'].apply(lambda x:x.split())
tokenized_tweet
```

```python
#stemmatize all words in a tweet
from nltk.stem.porter import *
stemmer = PorterStemmer()
tokenized_tweet = tokenized_tweet.apply(lambda x: [stemmer.stem(i) for i in x]) 
tokenized_tweet
```

```python
#tying stemmatized words back to sentence
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ' '.join(tokenized_tweet[i]) #concat all words in one sentence again
df['clean_tweet'] = tokenized_tweet
df['clean_tweet']
```

```python
#Creation of wordcloud
all_words = ' '.join([text for text in df['clean_tweet']])
from wordcloud import WordCloud
wordcloud = WordCloud(width=800, height=500, random_state=21, max_font_size=110).generate(all_words)
plt.figure(figsize = (10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

### Sentiment Analysis

```python
pip install textblob
```

```python
from textblob import TextBlob
```

```python
#Creating a function to be used as a dictionary
def get_sentiment(text):
    blob = TextBlob(text)
    sentiment_polarity = blob.sentiment.polarity
    sentiment_subjectivity = blob.sentiment.subjectivity
    if sentiment_polarity > 0:
        sentiment_label = 'Positive'
    elif sentiment_polarity < 0:
        sentiment_label = 'Negative'
    else:
        sentiment_label = 'Neutral'
    result = {'polarity':sentiment_polarity,
             'subjectivity':sentiment_subjectivity,
             'sentiment':sentiment_label}
    return result
```

```python
#Testing the dictionary
ex1 = df['clean_tweet'].iloc[0]
```

```python
get_sentiment(ex1)
```

```python
df['sentiment_results'] = df['clean_tweet'].apply(get_sentiment)
```

```python
#Preview of the results
df['sentiment_results']
```

```python
#appending the three columns into the data
df = df.join(pd.json_normalize(df['sentiment_results']))
```

### Using Multinomial Naive Bayes

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(stop_words = 'english')
```

```python
all_features = vectorizer.fit_transform(df['clean_tweet'])
```

```python
all_features.shape
```

```python
vectorizer.vocabulary_
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(all_features, df['sentiment'], test_size=0.2, random_state=88)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
classifier = MultinomialNB()
```

```python
classifier.fit(X_train, y_train)
```

```python
correct_tweet_prediction = (y_test == classifier.predict(X_test)).sum()
```

```python
print(f'{correct_tweet_prediction} tweets predicted correctly')
```

```python
incorrect_tweet = y_test.size - correct_tweet_prediction
```

```python
print(f'{incorrect_tweet} tweets predicted incorrectly')
```

```python
#Accuracy
classifier.score(X_test, y_test)
```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
```

```python
recall_score(y_test, classifier.predict(X_test))
```

```python
precision_score(y_test, classifier.predict(X_test))
```
